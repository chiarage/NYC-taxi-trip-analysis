7. Cluster the pickup time of the day into 30-minute intervals (e.g. from 02:00 to 02:30)

```python
def midhour(dataset):
    minute = dataset['tpep_pickup_datetime'].minute # Extract the minute value 
    hour = dataset['tpep_pickup_datetime'].hour # Extract the hour value 
    if minute >= 0 and minute < 30: #first half of the hour
        return f"from {hour}:00:00 to {hour}:30:00"  # Return the interval in string format
    else:  #second half of the hour
        return f"from {hour}:30:00 to {hour + 1}:00:00"  # Return the interval in string format
data['mid_hour_time']=data.apply(midhour, axis=1) #apply the 'midhour' function to each row (axis=1) to generate the new column 'mid_hour_time'
```
